```python
import json
import matplotlib.pyplot as plt

experiment_folder = '/home/asd/Project/BridgeDetection/Faster-RCNN-Trainer/output'

def load_json_arr(json_path):
    lines = []
    with open(json_path, 'r') as f:
        for line in f:
            lines.append(json.loads(line))
    return lines

experiment_metrics = load_json_arr(experiment_folder + '/metrics.json')

# plt.xticks(range(0, 5))
step = 80000
plt.figure(figsize=(16,8)) #整个现实图（框架）的大小
plt.plot(
    [x['iteration'] for x in experiment_metrics if x['iteration'] < step], 
    [x['total_loss'] for x in experiment_metrics if x['iteration'] < step])
plt.plot(
    [x['iteration'] for x in experiment_metrics if 'val_loss_laji_test' in x and  x['iteration'] < step], 
    [x['val_loss_laji_test'] for x in experiment_metrics if 'val_loss_laji_test' in x and x['iteration'] < step])
plt.legend(['total_loss', 'validation_loss'], loc='upper left')
plt.show()
```
